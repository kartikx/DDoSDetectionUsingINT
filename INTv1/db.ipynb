```python
from influxdb import InfluxDBClient
```

```python
client = InfluxDBClient(host='localhost', port=8086)
print(client.get_list_database())
```

```python
client.drop_database('flowDatabase')
```

```python
json_body = [
    {
        "measurement": "flowTable",
        "tags": {
            "srcIP": "10.0.0.1",
            "dstIP": "10.0.0.2"
        },
        "time": "2018-03-28T8:01:00Z",
        "fields": {
            "flowLatency": 1,
            "hopLatency": 2,
            "queueOccupancy": 3
        }
    }
]
```

```python
client.write_points(json_body)
```

```python
result = client.query("SELECT * from flowTable WHERE flowKey !~ /69.*/", database="flowDatabase")
output = list(result.get_points(measurement="flowTable"))

print(len(output))
for res in output:
    print(res)
```
